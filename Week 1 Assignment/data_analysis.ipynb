# Shopping Trends Data Analysis
**Objective:** Learn Python basics and perform basic data exploration and cleaning using Pandas.

**Steps:**
1. Load a CSV dataset into a Pandas DataFrame.
2. Explore data (head/tail, shape, columns, data types).
3. Handle missing values (identify, fill/drop).
4. Perform basic operations (filter rows, select columns).
5. Remove duplicates.
6. Create a derived column (`total_amount` = price * quantity).
7. Save the cleaned dataset as a new CSV file.

### Download the Dataset
Ensure you have downloaded the dataset from Kaggle: [Shopping Trends Dataset](https://www.kaggle.com/datasets/anvitkumar/shopping-dataset)
Place the `shopping_trends.csv` file in the same directory as this notebook.

```python
import pandas as pd
import numpy as np
import os
```

### 1. Load Dataset

```python
file_path = 'shopping_trends.csv'

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print('Dataset loaded successfully!')
else:
    print(f'Error: {file_path} not found. Please download it and place it in the same directory.')
    # Fallback to mock data for demonstration if file is missing
    print('\nUsing mock data for demonstration purposes...')
    data = {
        'Customer ID': [1, 2, 3, 4, 5, 5], 
        'Age': [25, np.nan, 35, 45, 22, 22], 
        'Item Purchased': ['Sweater', 'Jeans', 'Sneakers', 'Hat', 'Sweater', 'Sweater'],
        'Purchase Amount (USD)': [50, 80, 120, 15, 50, 50],
        'Quantity': [1, 2, 1, 3, 1, 1]
    }
    df = pd.DataFrame(data)
```

### 2. Explore Data

```python
print("\n--- FIRST 5 ROWS ---")
display(df.head())

print("\n--- LAST 5 ROWS ---")
display(df.tail())

print("\n--- DATASET SHAPE ---")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

print("\n--- COLUMN NAMES ---")
print(list(df.columns))

print("\n--- DATA TYPES ---")
print(df.dtypes)

print("\n--- DATASET INFO ---")
df.info()
```

### 3. Handle Missing Values

```python
print("--- MISSING VALUES BEFORE CLEANING ---")
print(df.isnull().sum())

# Fill numerical columns with mean
numeric_cols = df.select_dtypes(include=np.number).columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())

# Fill categorical columns with mode
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])

print("\n--- MISSING VALUES AFTER CLEANING ---")
print(df.isnull().sum())
```

### 4. Basic Operations

```python
# Select Columns
print("--- SELECTED COLUMNS (Customer ID, Age, Item Purchased) ---")
selected_data = df[['Customer ID', 'Age', 'Item Purchased']]
display(selected_data.head())

# Filter Rows
print("\n--- CUSTOMERS AGE > 30 ---")
filtered_data = df[df['Age'] > 30]
display(filtered_data.head())
```

### 5. Remove Duplicates

```python
print(f"ROWS BEFORE REMOVING DUPLICATES: {len(df)}")
df = df.drop_duplicates()
print(f"ROWS AFTER REMOVING DUPLICATES: {len(df)}")
```

### 6. Create Derived Column

```python
# If 'Quantity' doesn't exist in the actual dataset, create it for demonstration
if 'Quantity' not in df.columns:
    df['Quantity'] = 1  

# Create total_amount
df['total_amount'] = df['Purchase Amount (USD)'] * df['Quantity']

print("--- DERIVED COLUMN CREATED ---")
display(df[['Purchase Amount (USD)', 'Quantity', 'total_amount']].head())

print("\n--- SUMMARY STATISTICS ---")
display(df.describe())

print("\nTOTAL CUSTOMERS:", df['Customer ID'].nunique())
print("TOTAL SALES (USD): $", df['total_amount'].sum())
```

### 7. Save Cleaned Dataset & Summary

```python
# Save to CSV
cleaned_file_path = "cleaned_shopping_dataset.csv"
df.to_csv(cleaned_file_path, index=False)
print(f"CLEANED DATASET SAVED AS '{cleaned_file_path}'")

# Final Project Summary
print("\n" + "="*50)
print("PROJECT SUMMARY")
print("="*50)
print("""
1. Loaded the Shopping Trends dataset.
2. Explored data using head(), tail(), shape, columns, and data types.
3. Checked and handled missing values.
4. Performed filtering and column selection.
5. Removed duplicate records.
6. Created a derived column: total_amount.
7. Generated summary statistics.
8. Saved the cleaned dataset as cleaned_shopping_dataset.csv.
""")
```
